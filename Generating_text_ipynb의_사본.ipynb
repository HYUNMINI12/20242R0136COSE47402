```python
def build_faiss_index(monuments):
    all_indices = {}
    all_splited_docs = {}

    for monument, url in monuments.items():
        # 텍스트 로드
        response = requests.get(url)
        if response.status_code != 200:
            print(f"Failed to fetch URL: {url}")
            continue

        soup = BeautifulSoup(response.text, 'html.parser')
        paragraphs = soup.find_all('p')
        text = "\n".join([para.get_text() for para in paragraphs])

        if not text.strip():
            print(f"No content extracted for {monument}")
            continue

        # 텍스트 분할
        text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
        splited_docs = text_splitter.split_text(text)
        all_splited_docs[monument] = splited_docs

        if not splited_docs:
            print(f"No documents found for {monument} after splitting.")
            continue

        # FAISS 인덱스 생성
        embeddings = embedding_model.encode(splited_docs, batch_size=10)
        dimension = embeddings.shape[1]
        index = faiss.IndexFlatL2(dimension)
        index.add(embeddings)

        all_indices[monument] = index

    return all_indices, all_splited_docs

faiss_indices, all_splited_docs = build_faiss_index(monuments)
```

```python
# 이미지로 특정 유물 식별
def identify_monument(image_path):
    image = Image.open(image_path).convert("RGB")
    image_input = preprocess_clip(image).unsqueeze(0).to(device)

    text_inputs = list(monuments.keys())
    text_embeddings = clip.tokenize(text_inputs).to(device)

    with torch.no_grad():
        image_features = model_clip.encode_image(image_input)
        text_features = model_clip.encode_text(text_embeddings)

        image_features /= image_features.norm(dim=-1, keepdim=True)
        text_features /= text_features.norm(dim=-1, keepdim=True)

        similarity = image_features @ text_features.T
        best_match_idx = similarity.argmax().item()

    return text_inputs[best_match_idx]
```

```python
# ChatGPT API를 통한 텍스트 생성
def generate_monument_info_by_ChatGPT(monument_name):
    query = f"Provide a historical background and the establishment year of {monument_name}."
    splited_docs = all_splited_docs.get(monument_name, [])
    faiss_index = faiss_indices.get(monument_name, None)

    if not splited_docs or faiss_index is None:
        raise ValueError(f"No documents or index found for monument: {monument_name}")

    # FAISS 검색
    query_embedding = embedding_model.encode([query])
    distances, indices = faiss_index.search(np.array(query_embedding), k=3)

    # 유효한 인덱스 필터링
    valid_indices = [i for i in indices[0] if i < len(splited_docs)]
    if not valid_indices:
        raise ValueError(f"No valid documents found for monument: {monument_name}")

    # 검색된 문서 가져오기
    relevant_docs = [splited_docs[i] for i in valid_indices]
    context = "\n\n".join(relevant_docs)

    # OpenAI ChatGPT API 호출
    messages = [
        {"role": "system", "content": "You are a knowledgeable assistant providing detailed historical information."},
        {"role": "user", "content": f"Using the following context, provide a historical background and the establishment year of {monument_name}:\n\n{context}"}
    ]

    response = openai.ChatCompletion.create(
        model="gpt-4o",  # GPT-4 모델 사용 (또는 "gpt-3.5-turbo")
        messages=messages,
        max_tokens=500,
        temperature=0.7
    )

    return response['choices'][0]['message']['content'].strip()
```

```python
def generate_monument_info_by_LLaMA(monument_name):
    query = f"Provide a historical background and the establishment year of {monument_name}."
    splited_docs = all_splited_docs.get(monument_name, [])
    faiss_index = faiss_indices.get(monument_name, None)

    if not splited_docs or faiss_index is None:
        raise ValueError(f"No documents or index found for monument: {monument_name}")

    # FAISS 검색
    query_embedding = embedding_model.encode([query])
    distances, indices = faiss_index.search(np.array(query_embedding), k=3)

    print(f"Distances: {distances}")
    print(f"Indices: {indices}")

    # 유효한 인덱스 필터링
    valid_indices = [i for i in indices[0] if i < len(splited_docs)]
    if not valid_indices:
        print(f"Splited Docs Length: {len(splited_docs)}")
        raise ValueError(f"No valid documents found for monument: {monument_name}")

    # 검색된 문서 가져오기
    relevant_docs = [splited_docs[i] for i in valid_indices]
    context = "\n\n".join(relevant_docs)

    # Llama를 통한 텍스트 생성
    prompt = (
        f"Using the following context, write a well-crafted and natural narrative to answer the Question. \n\n"
        f"{context}\n\n"
        f"Question: {query}\n\n"
    )

    answer = qa_pipeline(
        prompt,
        max_new_tokens=500,
        num_return_sequences=1,
        top_p=0.9,
        temperature=0.7,
        do_sample=True,
        truncation=True
    )[0]['generated_text']

    return answer.strip()
```

```python
# GPT에 의한 처리 함수
def process_image_by_GPT(image_path):
    monument_name = identify_monument(image_path)
    print(f"Identified Monument: {monument_name}")
    monument_info = generate_monument_info_by_ChatGPT(monument_name)
    return monument_info
```

```python
# LLaMA에 의한 처리 함수
def process_image_by_LLaMA(image_path):
    monument_name = identify_monument(image_path)
    print(f"Identified Monument: {monument_name}")
    monument_info = generate_monument_info_by_LLaMA(monument_name)
    return monument_info
```
